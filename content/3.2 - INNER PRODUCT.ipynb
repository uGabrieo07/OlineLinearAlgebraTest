## Inner Product

Vector norm does not provide information of angles between vectors;
defining inner product will help us close this gap.

<span id="def:inner-product" label="def:inner-product"></span> Let
$\mathcal{U}$ be a vector space defined over the field $\mathbb{F}$,
where $\mathbb{F}$ can be either the field of real numbers $\mathbb{R}$
or the field of complex numbers $\mathbb{C}$. [1] Let $\mathbf{u}$ and
$\mathbf{v}$ belong to $\mathcal{U}$. The *inner product* between
$\mathbf{u}$ and $\mathbf{v}$, denoted herein as
$<\mathbf{u},\mathbf{v}>$, is a function
$<\cdot,\cdot>:\mathcal{U}\times\mathcal{U}\rightarrow\mathbb{F}$ which
satisfies the following properties:

-   $<\mathbf{u},\mathbf{u}>\ge 0$ with equality if, and only if,
    $\mathbf{u}=0$.

-   $<\mathbf{u}+\mathbf{w},\mathbf{v}>=<\mathbf{u},\mathbf{v}>+<\mathbf{w},\mathbf{v}>$,
    $\mathbf{w}\in\mathcal{U}$.

-   $<\alpha\mathbf{u},\mathbf{v}>=\alpha<\mathbf{u},\mathbf{v}>$,
    $\alpha\in\mathbb{F}$.

-   $<\mathbf{u},\mathbf{v}>=(<\mathbf{v},\mathbf{u}>)^*$, where the
    superscript $(\cdot)^*$ denotes complex conjugation.

In case we feel tempted to define an inner product for vector spaces
over arbitrary fields, i.e., other than $\mathbb{R}$ or $\mathbb{C}$, we
will stumble over positivity. There will exist a nonzero vector
$\mathbf{u}$ such that $<\mathbf{u},\mathbf{u}>=0$.

For a given inner product defined in a vector space $\mathcal{U}$,
$(<\mathbf{u},\mathbf{u}>)^{1/2}$ is a vector norm. The vector norm
$\|\mathbf{u}\|=(<\mathbf{u},\mathbf{u}>)^{1/2}$ is said to derive from
an inner product.

Some authors call Euclidean any norm derived from an inner product
defined in a vector space over $\mathbb{R}$ satisfying the properties
stated in
Definition <a href="#def:inner-product" data-reference-type="ref"
data-reference="def:inner-product">[def:inner-product]</a> \[@Gant1990\].
However, we will refer to Euclidean norm only the $l_2$ norm defined in
real vector spaces.

<span id="ex:1.6" label="ex:1.6"></span> Let
$f(\mathbf{u},\mathbf{v})=<\mathbf{u},\mathbf{v}>:\mathbb{R}^2\times\mathbb{R}^2\rightarrow\mathbb{R}$
be defined as
$$<\mathbf{u},\mathbf{v}>=\mathbf{v}^{\rm T}\mathbf{A}\mathbf{u},\ \text{where}\ \mathbf{A}=\left[\begin{matrix}1&1\\1&2\end{matrix}\right]$$
We shall verify that $f(\mathbf{u},\mathbf{v})$ is an inner product.
Indeed, the properties outlined in
Definition <a href="#def:inner-product" data-reference-type="ref"
data-reference="def:inner-product">[def:inner-product]</a> are all
satisfied:

-   Expanding $f(\mathbf{u},\mathbf{u})$, we obtain
    $<\mathbf{u},\mathbf{u}>=u_1^2+2u_1u_2+2u_2^2=(u_1+u_2)^2+u_2^2\ge0$.
    Therefore $f(\mathbf{u},\mathbf{u})$ is nonnegative and equal to
    zero if, and only if, $u_1=u_2=0$.

-   We can use the definition of linear transformation to verify that
    $f(\mathbf{u}+\mathbf{w},\mathbf{v})=f(\mathbf{u},\mathbf{v})+f(\mathbf{w},\mathbf{v})$
    and that
    $f(\alpha\mathbf{u},\mathbf{v})=\alpha f(\mathbf{u},\mathbf{v})$,
    $\alpha\in\mathbb{R}$.

-   We can either expand $f(\mathbf{v},\mathbf{u})$ to verify that the
    result is equal to $f(\mathbf{u},\mathbf{v})$, or we can use the
    fact that a real scalar is equal to its transpose, therefore
    $f(\mathbf{u},\mathbf{v})=\mathbf{v}^{\rm T}\mathbf{A}\mathbf{u}=\mathbf{u}^{\rm T}\mathbf{A}^{\rm T}\mathbf{v}$,
    together with the fact that $\mathbf{A}$ is a symmetric matrix,
    therefore $\mathbf{A}=\mathbf{A}^{\rm T}$, to obtain
    $f(\mathbf{u},\mathbf{v})=\mathbf{v}^{\rm T}\mathbf{A}\mathbf{u}=\mathbf{u}^{\rm T}\mathbf{A}^{\rm T}\mathbf{v}=\mathbf{u}^{\rm T}\mathbf{A}\mathbf{v}=f(\mathbf{v},\mathbf{u})$.
    Therefore $f(\mathbf{u},\mathbf{v})$ is symmetric.

The reader should redo the example above for other matrices, e.g.,
$$\mathbf{A}=\left[\begin{matrix}1&1\\2&1\end{matrix}\right],\ \text{or}\ \mathbf{A}=\left[\begin{matrix}3&1\\2&1\end{matrix}\right]$$
For both matrices above, one or more conditions fail and the quadratic
function
$f(\mathbf{u},\mathbf{v})=\mathbf{v}^{\rm T}\mathbf{A}\mathbf{u}$ cannot
be considered an inner product.

(Cauchy-Schwarz inequality) Let $<\cdot,\cdot>$ be an inner product
defined in a vector space $\mathcal{U}$ over the field $\mathbb{F}$
(either $\mathbb{R}$ or $\mathbb{C}$) and let
$\mathbf{u},\ \mathbf{v}\in\mathcal{U}$. The following inequality holds:
$$|<\mathbf{u},\mathbf{v}>|^2\le<\mathbf{u},\mathbf{u}><\mathbf{v},\mathbf{v}>$$
with equality if, and only if, $\mathbf{u}=\alpha\mathbf{v}$,
$\alpha\in\mathbb{F}$.

(Parallelogram equality) Let $\|\cdot\|$ be a norm derived from an inner
product, i.e., $\|\cdot\|=<\cdot,\cdot>^{1/2}$, defined in a vector
space $\mathcal{U}$ over the field $\mathbb{F}$. For any such norm, the
parallelogram equality is satisfied, i.e.,
$$\frac{1}{2}\left(\|\mathbf{u}+\mathbf{v}\|^2+\|\mathbf{u}-\mathbf{v}\|^2\right)=\|\mathbf{u}\|^2+\|\mathbf{v}\|^2$$

*Proof.* The proof is simple and follows from the distributive property
of inner products: $$\begin{split}
      \|\mathbf{u}+\mathbf{v}\|^2&=<\mathbf{u}+\mathbf{v},\mathbf{u}+\mathbf{v}>=<\mathbf{u},\mathbf{u}>+<\mathbf{u},\mathbf{v}>+<\mathbf{v},\mathbf{u}>+<\mathbf{v},\mathbf{v}>\\
      \|\mathbf{u}-\mathbf{v}\|^2&=<\mathbf{u}-\mathbf{v},\mathbf{u}-\mathbf{v}>=<\mathbf{u},\mathbf{u}>-<\mathbf{u},\mathbf{v}>-<\mathbf{v},\mathbf{u}>+<\mathbf{v},\mathbf{v}>
  \end{split}$$ 

[1] We opted to present here the more complete definition of inner
product, defined over either $\mathbb{R}$ or $\mathbb{C}$, despite our
decision to defer linear algebra over the complex field to the more
advanced chapters. We believe this serves better if the definition is
required in other contexts. If necessary, the reader may wish to
consider only the field of real numbers.

### Examples

A very useful tool in physics and engineering is named Fourier Series. Essentially, it is a linear combination of sines and cossines that seeks to represent a periodic signal. Periodic signals are everywhere: the electrical energy that reaches our homes is a sinusoidal function, radio stations are usually identified by the frequency with which you must syncronise in order to listen. It is interesting, therefore, to represent such signals as sums of sines and cosines, since these are well known periodic functions. 

As it was said before, Fourier Series seeks to represent a periodic signal. A continuous signal is periodic if it obeys the following equation: $x(t)=x(t+T_0)$ , when  $T_0$  is its fundamental period. 

It is basic knowledge that every periodic function has a frequency, which is calculated with: 

$$f_0=\frac{1}{T_0}  ,  w_0=2πf_0$$        

Multiples of the values found with the formulas above are known as harmonics. 

Generally speaking, we know that a linear combination is a sum of elements (which may or may not be multiplied by scalars) that results in another element. The elements involved in such sum typically form a basis, which is sub-space in a vector space that spans the entire vector space and that constitutes solely of linearly independent vectors. This is exactly what we observe in Fourier Series: 
$$x(t)=a_0+\sum_{n=0}^{\infty} a_n cos(n w_0 t) + \sum_{n=1}^{\infty} b_n sin(n w_0 t)$$  

We can see that 1 (which multiplies $a_0$),  $sin(nw0t)$, $cos(nw0t)$  are linearly independent. 
You cannot obtain any of the three elements above simply by adding or multiplying the other elements by scalars. That means they constitute a basis. 

Let the following integral:  $\int_{P} f(t)g(t) \,dx$   For a period of time P. 

This integral is very interesting. If we apply it in the following situations: 
  $$ \int_{P} cos(mt)cos(nt)dt = \left\{ 
    \begin{array}{l}
        \pi,  m = n &\\
        0,  m ≠ n
    \end{array} 
\right. 
$$

  $$\int_{P} sin(mt)sin(nt)dt = \left\{ 
    \begin{array}{l}
        \pi,  m = n &\\
        0,  m ≠ n
    \end{array} 
\right.$$
  $$\int_{P} sin(mt)cos(nt) \,dt = 0 $$ 
  $$\int_{P} 1 \,dt = 2\pi $$
  
 Dividing the result by P, we can see that the integral only returns 1 when the frequencies are the same for both sinusoids. For the value 1, it returns 2. Otherwise, it is always zero. Since sines and cosines are symmetric functions, if we apply the integral in the product between those functions and 1, we will also obtain zero. 
 Analyzing it a little bit further, it is possible to associate this integral with an inner product. It is linear, symmetric and positive definite. 
 We know that an inner product between two structures returns 0 when said structures are orthogonal. 
 Thus, we prove that the basis used in Fourier Series is orthogonal, which makes everything much easier. 
 In summary, the inner product in the integral between two functions through a period P, later divided by such period P.

 You might ask yourself how the coefficients that multiply the elements the basis are found. Once again, we use the inner product to define the coefficients: 
  $$a_0=\frac{1}{P} \int_{P}x(t)dt$$   $$a_n=\frac{2}{P} \int_{P}x(t)cos(n w_0 t)dt$$   $$bn=\frac{2}{P}\int_{P} x(t)sen(n w_0 t)dt$$  
 Generally speaking, the inner product reveals what two structures have in common. Therefore, the application above is a projection that reveals the similarity between the function that we intend to rewrite and the elements of the basis. That makes sense, because in order to construct a signal from a new basis, we need to know what the signal and that basis have in common. 
 In order to find those coefficients, we choose the value P to be equal to that of the period of the signal x(t).

According to the definition, the linear combination is an infinite sum of scaled sines and cosines. This means that we are using an infinite basis to construct the periodic signal. What happens if we limit the amount of sines and cosines in the sums? Change the value of the variable n to see what happens and compare that to the original signal. 

What is the effect observed by increasing the value of n? 

```python
import numpy as np
from scipy import signal as signal
import matplotlib.pyplot as plt

w = 1000
t = np.linspace(0, 0.03, 100, endpoint=False)
n = 2
harmonics = range(n)
for inteiro in harmonics: 
    soma = 0
    for indice in range(1, inteiro+1):
        serie = ((-1**indice)*(np.sin((indice*t*w))))/(indice)
        soma += serie
    entrada = soma*2/np.pi

plt.subplot(1, 2, 2)
plt.plot(t, entrada, color= "blue", alpha=0.6, linewidth=2)
plt.grid()
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(f"Fourier series with {n} elements")

plt.subplot(1,2,1)
plt.plot(t, signal.sawtooth(w*t), color= "red", alpha=0.6, linewidth=2)
plt.grid()
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title("Sawtooth signal")
plt.show()
```
