<span style="color:#9c4be9; font-size:25px">What is JupyterLite?</span>

<span style="color:black; font-size:18px"> This project is made public and accessible with the support of JupyterLite. It is JupyterLab distribution that creates an environment to create Jupyter Notebooks using WebAssembly. Essentialy, it creates a Jupyter environment compatible with a browser, that can be accessed through a URL.<br>
As you have probably already seen, the home page looks something like this:</span>

<span style="color:black; font-size:18px"> In the left side of the screen, you can see all of the available notebooks in the project. By clicking on them, you will be able to visualize markdown text, as well as execute code with a Pyodide kernel. In that sense, you can create your own notebooks using JupyterLite, as well as edit preexisting one. The environment is entirely costumizable!</span>

<span style="color:black; font-size:18px">This part allows you to create new things from scratch. By clicking in the two icons below "Notebook", you will be redirected to an IDLE in which you can create your own Jupyter Notebooks. And by clicking in the icons below "Console', you will be redirected to an environment where you can run code interactively. In this project, it will not be necessary for you to create your own notebooks nor use the console; all of the necessary material is already available. </span>

<span style="color:black; font-size:18px">Yet, do not worry if you happen to make unwanted changes: a new, fresh environment is created every time you click on the URL. If you notice that, even clicking on the URL once again, the altered environment is still exhibited, it is likely that you are seeing the version of the environment saved in cache memory. To access the most updated release version, try accessing the URL in incognito mode!</span>
