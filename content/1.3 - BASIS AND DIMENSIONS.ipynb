# Bases and Dimension

Given a space $\mathcal{E}$, it may be quite useful to be able to
address (or to access) each and every element of $\mathcal{E}$ using a
set of unequivocal coordinates. Latitudes and longitudes provide such
coordinates for geographical positioning on earth‚Äôs surface. For a more
generic space, we may need to define a basis for the space in more
appropriate terms.

Given a space $\mathcal{E}$ defined over a field $\mathbb{F}$, a set of
elements $\{\mathbf{v}_i\}\subseteq\mathcal{E}$, $i=1,\ 2,\ \ldots,\ r$,
is a *basis* of $\mathcal{E}$ if the elements
$\mathbf{v}_i\in\{\mathbf{v}_i\}$ are linearly independent and any
element $\mathbf{u}\in\mathcal{E}$ can be written as a linear
combination of $\mathbf{v}_i$. Another way to put it is to say that the
elements $\mathbf{v}_i$, $i=1,\ 2,\ \ldots,\ r$ span the whole space
$\mathcal{E}$ and no other set of elements with less than $r$ elements
spans the space $\mathcal{E}$.

As many different coordinate systems may prove equally accurate for
determining position on the earth‚Äôs surface, many different bases may
span a given vector space.

<span id="ex:canonic_basis" label="ex:canonic_basis"></span> Let us
investigate the vector space $\mathbb{R}^2$. As previously defined, this
space can be described as the set
$$\mathbb{R}^2=\{\mathbf{v};\ \mathbf{v}=\left[v_1\ v_2\right]^{\rm T},\ v_1,\ v_2 \in \mathbb{R}\}$$
We may verify that any vector $\mathbf{v}\in\mathbb{R}^2$ can be written
as a linear combination of vectors $\mathbf{e}_1$ and $\mathbf{e}_2$
defined below:
$$\mathbf{e}_1=\left[\begin{matrix}1\\0\end{matrix}\right],\quad
\mathbf{e}_2=\left[\begin{matrix}0\\1\end{matrix}\right].$$ It is also
straightforward to verify that
$\mathbf{e}_1,\ \mathbf{e}_2\in\mathbb{R}^2$. Therefore the set of
vectors $\{\mathbf{e}_1,\ \mathbf{e}_2\}$ span $\mathbb{R}^2$. This
particular set of vectors forms the canonic basis of $\mathbb{R}^2$,
although it may not be so easy, at least at this moment, to ascertain
that no set with less than two elements span the whole space
$\mathbb{R}^2$. Many other sets with two vectors can form a basis of
$\mathbb{R}^2$. We will see further in the chapter how to change from
one basis to another.

<span id="ex:V_space_line" label="ex:V_space_line"></span> Another
interesting example we may construct is for a vector space
$\mathcal{V}\subset\mathbb{R}^2$ defined as
$$\mathcal{V}=\{\mathbf{v};\ \mathbf{v}=\alpha \left[1 \ 2\right]^{\rm T},\ \alpha\in\mathbb{R}\}$$
In this example, $\mathcal{V}$ is a line, as seen in

point along this line using any basis of $\mathbb{R}^2$ (after all,

In fact, any vector $\mathbf{v}=\alpha\left[1 \ 2\right]^{\rm T}$ is
itself a basis of $\mathcal{V}$.

The definition of basis implies that for every vector space there is a
set of vectors forming its basis. We have seen above that the number of
vectors in the basis is two for $\mathbb{R}^2$ and one for $\mathcal{V}$
in Example¬†<a href="#ex:V_space_line" data-reference-type="ref"
data-reference="ex:V_space_line">[ex:V_space_line]</a>. This number
denotes the dimension of these vector spaces.

Given a vector space $\mathcal{U}$, we define the *dimension* of the
vector space as the total number of vectors in its basis. A given vector
space may have infinite dimension, or it may have dimension equal to
zero.

```python
from PIL import Image 
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interactive_output
```

```python
escalar01 = widgets.FloatText(value = 1, description = 'Red of first plot:', disabled = False, style= {'description_width': 'initial'})
escalar02 = widgets.FloatText(value = 1, description = 'Black of first plot:', disabled = False, style= {'description_width': 'initial'})
escalar03 = widgets.FloatText(value = 1, description = 'Red of second plot:', disabled = False,style= {'description_width': 'initial'})
escalar04 = widgets.FloatText(value = 1, description = 'Black of second plot:', disabled = False, style= {'description_width': 'initial'})
formatted1 = widgets.VBox([escalar01, escalar02])
formatted2 = widgets.VBox([escalar03, escalar04])
formatted3 = widgets.HBox([formatted1, formatted2])

def setas1(escalar01, escalar02, escalar03, escalar04):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    plt.figure(figsize=(2, 2))
    ax1.arrow(0, 0, escalar01, 0, width = 0.01, color = 'red', length_includes_head = True, head_width = 0.03)
    ax1.arrow(0, 0, 0, escalar02, width = 0.01, color = 'black', length_includes_head = True, head_width = 0.03)
    ax1.arrow(0, 0, escalar01, escalar02, width = 0.01, length_includes_head = True, head_width = 0.03)
    ax1.grid()
    ax2.arrow(0, 0, 4*escalar03, 3*escalar03, width = escalar03*(0.03), color = 'red', length_includes_head = True, head_width = escalar03*(0.08))
    ax2.arrow(0, 0, 0, 2*escalar04, width = escalar03*(0.03), color = 'black', length_includes_head = True, head_width = escalar03*(0.08))
    ax2.arrow(0, 0, 4*escalar03, 3*escalar03 + 2*escalar04, width = escalar03*(0.03), length_includes_head = True, head_width = escalar03*(0.08))
    ax2.grid()
    plt.show()
    return 0
    
verif = interactive_output(setas1, {'escalar01': escalar01, 'escalar02': escalar02, 
                                           'escalar03': escalar03, 'escalar04': escalar04})
display(verif, formatted3)
```

#### Applications in image processing

So far, we have analyzed primarily grayscale images. You might have asked yourself how colored images are constructed. In order to understand that we must recall concepts of the previous chapter. (1.2 if you wish to read it again)
Recall that a vector space is a algebraic structure with the following properties:

For addition:
-   Associativity: For any elements
    $\mathbf{u},\ \mathbf{v},\ \mathbf{w}\in\mathcal{E}$,
    $(\mathbf{u}+\mathbf{v})+\mathbf{w}=\mathbf{u}+(\mathbf{v}+\mathbf{w})$.

-   Commutativity: For any elements
    $\mathbf{u},\ \mathbf{v}\in\mathcal{E}$,
    $\mathbf{u}+\mathbf{v}=\mathbf{v}+\mathbf{u}$.

-   Identity: $\exists 0\in\mathcal{E}$ such that
    $0+\mathbf{u}=\mathbf{u}$ for any element
    $\mathbf{u}\in\mathcal{E}$.

-   Inverse: Any element $\mathbf{u}\in\mathcal{E}$ has a symmetric
    $-\mathbf{u}$ such that $\mathbf{u}+(-\mathbf{u})=0$.

For multiplication by a scalar:

-   Associativity: For any element $\mathbf{u}\in\mathcal{E}$ and
    scalars $x,\ y\in\mathbb{F}$, $(xy)\mathbf{u}=x(y\mathbf{u})$.

-   Identity: $\exists 1\in\mathbb{F}$ such that
    $1\mathbf{u}=\mathbf{u}$ for any element $\mathbf{u}\in\mathcal{E}$.

For addition and multiplication:

-   Distributivity of the multiplication by a scalar with respect to the
    vector addition: For any $x\in\mathbb{F}$ and
    $\mathbf{u},\ \mathbf{v}\in\mathcal{E}$,
    $x(\mathbf{u}+\mathbf{v})=x\mathbf{u}+x\mathbf{v}$.

-   Distributivity of the scalar addition with respect to the
    multiplication by a scalar: For any $x,\ y\in\mathbb{F}$ and
    $\mathbf{u}\in\mathcal{E}$,
    $(x+y)\mathbf{u}=x\mathbf{u}+y\mathbf{u}$.
    
Notice that, when defining these opperations, we do not define the structure of the elements, which means that a vector space may not be defined exclusively by the use of "traditional" forms of vectors, such as: $\mathbf{v}_1=\left[\begin{matrix}a\\b\end{matrix}\right]$, for example. A vector, therefore, in the context of vector spaces, can be understood as a structure that is defined by elements of a field. So, we can think of polynomials as being vectors, as well as functions, and, naturally, coordinates. We can also define vector spaces of matrices, which is the space in which we develop colored images. In order to visualize this easier, think of a vector space that is not visually seen by coordinates,but rather a set of matrices that follow the properties mentioned above for addition and scalar multiplication. 

In most applications, colored images are matrices defined in vector spaces with dimension 3 and fined over field $GF(256)$. This means that the basis to form a colored image is composed of three matrices. Each one of these matrices describe a color. In most applications, we have a matriz for tones of red, named R, a matrix for tones for green, named G, and a matrix for tones of blue, named B. Therefore, a colored image is the linear combination of a basis, in which each matrix represents a color. The pixels of an image are the elements of a matrix. Any color can be represented as the sum of a shade of red, a shade of green and a shade of blue. 

We can conclude that a higher dimension allows more information to be included. If we recall grayscale images, a grayscale image is not the result of a sum, it is defined by itself. A grayscale image is defined in a vector space with dimension equal to 1. By increasing the dimension, we can allow new information to be included. 

The example below seeks to allow visualization of the components of an image. Each of the venentian masks were drawn with a pure color, thus, with highest value (255) for a specific component. Therefore, the mask with the same color as a component will disappear and the other two will become black, since black represents the complete absence of said color. The background of the resulting image is white. This happens because white is the result of the highest value in all color components. 

In order to visualize the components, click on the log below after pressing the buttons.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown
```

```python
# Creating Output
output = widgets.Output()

# Opening the venentian masks image
masks = Image.open("venetianmasks.png")
matrizImagem = np.array(masks)

plt.imshow(matrizImagem)
plt.axis('off')
plt.show()

# Creating copies for the R, G and B images
matrizImagemR = np.copy(matrizImagem)
matrizImagemG = np.copy(matrizImagem)
matrizImagemB = np.copy(matrizImagem)

# Isolating each component
matrizImagemR[:, :, 1:] = 0  # Mant√©m apenas o canal R
matrizImagemG[:, :, [0, 2]] = 0  # Mant√©m apenas o canal G
matrizImagemB[:, :, :2] = 0  # Mant√©m apenas o canal B


# Function to show each of the components
def showComponent(component):
    with output:
        clear_output(wait=True)
        plt.figure(figsize=(5, 5))
        if component == 'R':
            plt.imshow(matrizImagemR)
        elif component == 'G':
            plt.imshow(matrizImagemG)
        elif component == 'B':
            plt.imshow(matrizImagemB)
        plt.axis("off")
        plt.show()

display(output)
interact(showComponent, component=widgets.ToggleButtons(options=['R', 'G', 'B'], description='Component:'))
```
