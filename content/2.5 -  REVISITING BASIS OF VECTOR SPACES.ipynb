## Revisiting Bases of Linear Vector Spaces

An interesting and simple application of a linear operator in a vector
space is to change the basis used to represent a given element in the
space. We represent an element $\mathbf{u}$ in a vector space
$\mathcal{U}$ as an $n$-tuple
$\mathbf{u}=[u_1\ u_2\ \cdots\ u_n]^{\rm T}$, and it is perhaps natural
to associate the $n$-tuple with the canonic basis, as was done in
Example <a href="#ex:canonic_basis" data-reference-type="ref"
data-reference="ex:canonic_basis">[ex:canonic_basis]</a> for
$\mathbb{R}^2$. However arguably natural, the choice of the canonic
basis $\{\mathbf{e}_i\}$ is only one among many possibilities. The same
element of the space $\mathcal{U}$ may be written in different bases as
different linear combinations of the respective basis elements. Given
some arbitrary basis $\{\mathbf{v}_i\}$, a particular element
$\mathbf{u}\in\mathcal{U}$ is represented as
$\alpha_1 \mathbf{v}_1 + \alpha_2 \mathbf{v}_2 + \ldots + \alpha_n \mathbf{v}_n$,
whereas in some other basis $\{\mathbf{w}_i\}$ the same element is
represented as
$\beta_1 \mathbf{w}_1 + \beta_2 \mathbf{w}_2 + \ldots + \beta_n \mathbf{w}_n$.
Using the matrix notation, we may say that
$$[\begin{matrix}\mathbf{v}_1&\mathbf{v}_2&\cdots&\mathbf{v}_n\end{matrix}]\left[\begin{matrix}\alpha_1\\\alpha_2\\\vdots\\\alpha_n\end{matrix}\right]
=
[\begin{matrix}\mathbf{w}_1&\mathbf{w}_2&\cdots&\mathbf{w}_n\end{matrix}]\left[\begin{matrix}\beta_1\\\beta_2\\\vdots\\\beta_n\end{matrix}\right]$$
or, equivalently,
$$\mathbf{V}\left[\begin{matrix}\alpha_1\\\alpha_2\\\vdots\\\alpha_n\end{matrix}\right]=
\mathbf{W}\left[\begin{matrix}\beta_1\\\beta_2\\\vdots\\\beta_n\end{matrix}\right]$$
Moving from one representation to another, i.e., changing bases, is
always a bijection, therefore induced by a square invertible matrix. If
$\mathbf{V}$ and $\mathbf{W}$ are the matrices whose columns are the
elements of the bases $\{\mathbf{v}_i\}$ and $\{\mathbf{w}_i\}$,
respectively, going from one representation to the other is a composite
linear transformation. If $\exists\mathbf{W}^{-1},\mathbf{V}^{-1}$,
then:
$$\mathbf{V}\left[\begin{matrix}\alpha_1\\\alpha_2\\\vdots\\\alpha_n\end{matrix}\right]
=
\mathbf{W}\left[\begin{matrix}\beta_1\\\beta_2\\\vdots\\\beta_n\end{matrix}\right]
  \Longrightarrow
\left[\begin{matrix}\beta_1\\\beta_2\\\vdots\\\beta_n\end{matrix}\right]=
\mathbf{W}^{-1}\mathbf{V}
\left[\begin{matrix}\alpha_1\\\alpha_2\\\vdots\\\alpha_n\end{matrix}\right]$$

In case $\mathbf{W}^{-1},\mathbf{V}^{-1}$ do not exist, then changing
basis is slightly more involved, but still always possible and induced
by a square invertible matrix. Let $\mathcal{U}\subseteq\mathbb{R}^m$ be
a vector space defined over the field $\mathbb{R}$, whose dimension is
$n\le m$. Let $\{\mathbf{v}_i\}$ and $\{\mathbf{w}_i\}$,
$i=1,\ \ldots,\ n$, be bases of $\mathcal{U}$. Each element of the basis
are $m$-tuples, but they also must belong to $\mathcal{U}$, i.e.,
$\mathbf{v}_i\in\mathcal{U}$ and $\mathbf{w}_i\in\mathcal{U}$, for
$i=1,\ 2\,\ldots,\ n$. Therefore we can write the very elements of one
basis using another basis: $$\begin{split}
    \mathbf{v}_1 &= t_{11}\mathbf{w}_1 + t_{21}\mathbf{w}_2 + \cdots + t_{n1}\mathbf{w}_n\\
    \mathbf{v}_2 &= t_{12}\mathbf{w}_1 + t_{22}\mathbf{w}_2 + \cdots + t_{n2}\mathbf{w}_n\\
    \vdots & \\
    \mathbf{v}_n &= t_{1n}\mathbf{w}_1 + t_{2n}\mathbf{w}_2 + \cdots + t_{nn}\mathbf{w}_n
\end{split}$$ or, equivalently, $$\begin{split}
&   \left[\begin{matrix}
    \mathbf{v}_1 & \mathbf{v}_2 & \cdots & \mathbf{v}_n
    \end{matrix}\right] =
    \left[\begin{matrix}
    \mathbf{w}_1 & \mathbf{w}_2 & \cdots & \mathbf{w}_n
    \end{matrix}\right]
    \left[\begin{matrix}
    t_{11} & t_{12} & \cdots & t_{1n}\\
    t_{21} & t_{22} & \cdots & t_{2n}\\
    \vdots &        & \ddots & \vdots\\ 
    t_{n1} & t_{n2} & \cdots & t_{nn}
    \end{matrix}\right]\\
&
    \mathbf{V}=\mathbf{W}\mathbf{T}
\end{split}$$ The *matrix of coordinate transformation*, $\mathbf{T}$,
can be obtained, e.g., as
$$\mathbf{T}=(\mathbf{W}^{\rm T}\mathbf{W})^{-1}\mathbf{W}^{\rm T}\mathbf{V}$$
As the columns of $\mathbf{W}$ are linearly independent, which is a
necessary condition for a basis, then the inverse
$(\mathbf{W}^{\rm T}\mathbf{W})^{-1}$ always exists. This can be
verified as follows: let $\mathbf{w}_i$ be the ith column of matrix
$\mathbf{W}$. The ith column of matrix $\mathbf{W}^{\rm T}\mathbf{W}$
is $\mathbf{W}^{\rm T}\mathbf{w}_i$ and we can write a linear
combination of all columns of $\mathbf{W}^{\rm T}\mathbf{W}$ as
$\sum_i\alpha_i\mathbf{W}^{\rm T}\mathbf{w}_i=\mathbf{W}^{\rm T}\sum\alpha_i\mathbf{w}_i$,
which is equal to zero if, and only if, $\alpha_i=0$,
$i=1,\ 2,\ \ldots,\ n$. Notice that
$\mathbf{W}^{\rm T}\mathbf{w}_i\neq\mathbf{0}$, $i=1,\
2,\ \ldots,\ n$ because at least the ith element of
$\mathbf{W}^{\rm T}\mathbf{w}_i$ is equal to $\sum_jw_{ij}^2$, which
cannot be zero for $\mathbf{w}_i\neq\mathbf{0}$. Therefore the columns
of $\mathbf{W}^{\rm T}\mathbf{W}$ are linearly independent. Using
Theorem <a href="#theo:injective-lin_ind_cols" data-reference-type="ref"
data-reference="theo:injective-lin_ind_cols">[theo:injective-lin_ind_cols]</a>,
we know that $\mathbf{W}^{\rm T}\mathbf{W}$ is injective and has a left
inverse. But as $\mathbf{W}^{\rm T}\mathbf{W}$ is symmetric, its left
inverse is equal to its right inverse. Therefore it is invertible.

We can also say that
$$\mathbf{V}\mathbf{T}^{-1}=\mathbf{W}\Rightarrow \mathbf{T}^{-1}=(\mathbf{V}^{\rm T}\mathbf{V})^{-1}\mathbf{V}^{\rm T}\mathbf{W}$$
The same argument used for the existence of
$(\mathbf{W}^{\rm T}\mathbf{W})^{-1}$ applies to the existence of
$(\mathbf{V}^{\rm T}\mathbf{V})^{-1}$. Notice that if the number of
elements in the vectors is the same as the dimension of the space,
$n=m$, then matrices $\mathbf{W}$ and $\mathbf{V}$ will be square and
invertible, therefore $\mathbf{T}=\mathbf{W}^{-1}\mathbf{V}$, or,
equivalently, $\mathbf{T}^{-1}=\mathbf{V}^{-1}\mathbf{W}$.

Now that we know how to represent elements of $\mathcal{U}$ using
different bases, let us investigate how linear operators in
$\mathcal{U}$ relate to a change of basis. Let
$\mathbf{A}:\mathcal{U}\rightarrow\mathcal{U}$ be a linear operator in
$\mathcal{U}$. Let and $\mathbf{x}$ and
$\mathbf{y}=\mathbf{A}\mathbf{x}$ be elements in $\mathcal{U}$. We can
write $\mathbf{x}$ using basis $\{\mathbf{v}_i\}$, or using basis
$\{\mathbf{w}_i\}$, as
$$\mathbf{x}=\sum_{i=1}^n\alpha_i\mathbf{v}_i=\mathbf{V}\left[\begin{matrix}\alpha_1\\\alpha_2\\\vdots\\\alpha_n\end{matrix}\right]$$
or
$$\mathbf{x}=\sum_{i=1}^n\beta_i\mathbf{w}_i=\mathbf{W}\left[\begin{matrix}\beta_1\\\beta_2\\\vdots\\\beta_n\end{matrix}\right]$$
Notice that $\mathbf{x}\in\mathcal{U}$ and
$\mathcal{U}\subseteq\mathbb{R}^m$, therefore $\mathbf{x}$ is
represented as an $m$-tuple. However, the coefficients
$\alpha_1,\ \ldots,\ \alpha_n$, or $\beta_1,\ \ldots,\ \beta_n$, for the
representation using basis $\{\mathbf{v}_i\}$, or $\{\mathbf{w}_i\}$,
respectively, are $n$-tuples. We may say the same thing with respect to
the response to the linear transformation, $\mathbf{y}$:
$\mathbf{y}\in\mathcal{U}$ is represented as an $m$-tuple, but the
coefficients which represent $\mathbf{y}$ using either basis
$\{\mathbf{v}_i\}$, or $\{\mathbf{w}_i\}$, are $n$-tuples. Remember that
the $m$-tuples are representations of $\mathbf{x}$ and $\mathbf{y}$
using the canonic basis
$\{\mathbf{e}_1,\ \mathbf{e}_2,\ \mathbf{e}_3\}$, which is not a basis
of $\mathcal{U}$.

Let
$\mathbf{x}_v=\left[\begin{matrix}\alpha_1&\alpha_2&\cdots&\alpha_n\end{matrix}\right]^{\rm T}$
and
$\mathbf{x}_w=\left[\begin{matrix}\beta_1&\beta_2&\cdots&\beta_n\end{matrix}\right]^{\rm T}$
be the vectors with the coefficients of $\mathbf{x}$ using basis
$\{\mathbf{v}_i\}$ and $\{\mathbf{w}_i\}$, respectively, and
$\mathbf{y}_v$ and $\mathbf{y}_w$ be the corresponding responses to the
linear operator $\mathbf{y}=\mathbf{A}\mathbf{x}$ using the same
respective basis. Therefore,
$$\mathbf{y}=\mathbf{V}\mathbf{y}_v=\mathbf{A}\mathbf{x}=\mathbf{A}\mathbf{V}\mathbf{x}_v$$
As $\mathbf{V}$ is not necessarily square, but in any case
$\mathbf{V}^{\rm T}\mathbf{V}$ is square and guaranteed invertible, we
can do
$$\mathbf{V}^{\rm T}\mathbf{y}=\mathbf{V}^{\rm T}\mathbf{V}\mathbf{y}_v=\mathbf{V}^{\rm T}\mathbf{A}\mathbf{V}\mathbf{x}_v$$
As a consequence $$\begin{split}
    \mathbf{y}_v&=\left(\mathbf{V}^{\rm T}\mathbf{V}\right)^{-1}\mathbf{V}^{\rm T}\mathbf{A}\mathbf{V}\mathbf{x}_v\\
          &=\mathbf{A}_v\mathbf{x}_v
\end{split}$$ where we have just defined the linear operator
$\mathbf{A}_v=\left(\mathbf{V}^{\rm T}\mathbf{V}\right)^{-1}\mathbf{V}^{\rm T}\mathbf{A}\mathbf{V}$.
We can similarly define
$\mathbf{A}_w=\left(\mathbf{W}^{\rm T}\mathbf{W}\right)^{-1}\mathbf{W}^{\rm T}\mathbf{A}\mathbf{W}$
for $\mathbf{y}_w=\mathbf{A}_w\mathbf{x}_w$. Here, as before, the
existence of the inverses
$\left(\mathbf{V}^{\rm T}\mathbf{V}\right)^{-1}$ and
$\left(\mathbf{W}^{\rm T}\mathbf{W}\right)^{-1}$ should be of no
concern, for the columns of $\mathbf{V}$ and $\mathbf{W}$ are linearly
independent.

As $\mathbf{V}=\mathbf{W}\mathbf{T}$ and
$(\mathbf{T}^{-1}\mathbf{W}^{\rm T}\mathbf{W}\mathbf{T})^{-1}=\mathbf{T}^{-1}(\mathbf{W}^{\rm T}\mathbf{W})^{-1}\mathbf{T}^{-{\rm T}}$,
we have $$\begin{split}
    \mathbf{A}_v&=\left(\mathbf{T}^{\rm T}\mathbf{W}^{\rm T}\mathbf{W}\mathbf{T}\right)^{-1}\mathbf{T}^{\rm T}\mathbf{W}^{\rm T}\mathbf{A}\mathbf{W}\mathbf{T}\\
          &=\mathbf{T}^{-1}\left(\mathbf{W}^{\rm T}\mathbf{W}\right)^{-1}\mathbf{W}^{\rm T}\mathbf{A}\mathbf{W}\mathbf{T}\\
          &=\mathbf{T}^{-1}\mathbf{A}_w\mathbf{T}
\end{split}$$ from where we realize that two matrices corresponding to
the same linear operator but in different bases are similar, according
to the following definition of similarity transformation.

<span id="def:sim_matrices" label="def:sim_matrices"></span> Two
matrices $\mathbf{A}_v$ and $\mathbf{A}_w$ are called *similar* if they
are related by a *similarity transformation*, i.e.,
$$
\mathbf{A}_v = \mathbf{T}^{-1} \mathbf{A}_w \mathbf{T}
$$

where $\mathbf{T}$ is nonsingular.

Let us revisit Example [ex:inverse](#ex:inverse) and investigate different bases for $\mathcal{U}$.

<span id="ex:inverse-revisited" label="ex:inverse-revisited"></span>

Let $\mathcal{U}$ be a vector space defined over the field $\mathbb{R}$:

$$
\mathcal{U} = \left\{ \mathbf{u} : \mathbf{u} = \left[\begin{matrix} \alpha \\ \beta \\ \alpha + \beta \end{matrix}\right], \ \alpha, \beta \in \mathbb{R} \right\}
$$One basis of $\mathcal{U}$ is, e.g.,
$$
\mathcal{V} = \left\{ \begin{bmatrix} 1 \\ 0 \\ 1 \end{bmatrix}, \ \begin{bmatrix} 0 \\ 1 \\ 1 \end{bmatrix} \right\}
$$

therefore, the dimension of $\mathcal{U}$ is $n = 2$, whereas the elements of the space $\mathcal{U}$ are vectors with $m = 3$ elements. Another basis of $\mathcal{U}$ is, e.g.,

$$
\mathcal{W} = \left\{ \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}, \ \begin{bmatrix} 1 \\ 1 \\ 2 \end{bmatrix} \right\}
$$

Using the fact that we can write the elements of $\mathcal{V}$ as a linear combination of the elements of $\mathcal{W}$, we obtain a matrix $\mathbf{T}$ which relates both bases as follows:

$$
\mathbf{V} = \mathbf{W} \mathbf{T} \Rightarrow 
\begin{bmatrix} 1 & 0 \\ 0 & 1 \\ 1 & 1 \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 2 & 1 \\ 3 & 2 \end{bmatrix} \mathbf{T}
$$
Solving for $\mathbf{T}$ and obtaining $\mathbf{T}^{-1}$ may require a
leap of faith from the reader at this point, even though the matrices
involved are small. However, the reader should have no difficulty to
verify that the relations are correct and to understand the mechanisms
involved when changing basis:
$\mathbf{T}=(\mathbf{W}^{\rm T}\mathbf{W})^{-1}\mathbf{W}^{\rm T}\mathbf{V}$
gives
$$\mathbf{T}=\left[\begin{matrix}-1&1\\2&-1\end{matrix}\right] \Leftrightarrow
  \mathbf{T}^{-1}=\left[\begin{matrix}1&1\\2&1\end{matrix}\right]$$ We
may say that a vector
$\mathbf{x}=\left[\begin{matrix}1&1&2\end{matrix}\right]^{\rm T}\in\mathcal{U}$
can be written with respect to basis $\mathcal{V}$ as
$$
\mathbf{x}_v = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

For the same linear operator defined in Example [ex:inverse](#ex:inverse), and repeated below for convenience:

$$
\mathbf{A} : \mathcal{U} \rightarrow \mathcal{U}, \qquad \mathbf{A} = \begin{bmatrix} 1 & 1 & 1 \\ 1 & 0 & 1 \\ 4 & 3 & 0 \end{bmatrix}
$$

the operation $\mathbf{A}\mathbf{x}$ yields
$$\mathbf{y}=\mathbf{A}\mathbf{x}=\left[\begin{matrix}4&3&7\end{matrix}\right]^{\rm T}$$
whose representation using basis $\mathcal{V}$ is
$$\mathbf{y}_v=\left[\begin{matrix}4&3\end{matrix}\right]^{\rm T}
  =\left(\mathbf{V}^{\rm T}\mathbf{V}\right)^{-1}\mathbf{V}^{\rm T}\mathbf{A}\mathbf{V}\mathbf{x}_v=\mathbf{A}_v\mathbf{x}_v$$
The linear operator $\mathbf{A}_v:\mathbb{R}^2\rightarrow\mathbb{R}^2$
is
$$\mathbf{A}_v=(\mathbf{V}^{\rm T}\mathbf{V})^{-1}\mathbf{V}^{\rm T}\mathbf{A}\mathbf{V}=\left[\begin{matrix}2&2\\2&1\end{matrix}\right]$$
For the basis $\mathcal{W}$, we have
$\mathbf{x}_w=\left[\begin{matrix}0&1\end{matrix}\right]^{\rm T}$,
$\mathbf{y}_w=\left[\begin{matrix}-1&5\end{matrix}\right]^{\rm T}$, and
$$\mathbf{A}_w=(\mathbf{W}^{\rm T}\mathbf{W})^{-1}\mathbf{W}^{\rm T}\mathbf{A}\mathbf{W}=\left[\begin{matrix}-2&-1\\8&5\end{matrix}\right]$$
Confirmation that $\mathbf{A}_v$ and $\mathbf{A}_w$ are similar is
straightforward: $$\mathbf{A}_w=\mathbf{T}\mathbf{A}_v\mathbf{T}^{-1}$$

As a final observation on the subject, we call the attention to the fact
that $\mathbf{A}$ operates on the elements of $\mathcal{U}$ written in
the canonic basis, whereas $\mathbf{A}_v$ and $\mathbf{A}_w$ operate on
the same elements, but written in the bases $\mathcal{V}$ and
$\mathcal{W}$, respectively. As we have mentioned before, it is
important to establish unequivocally the basis in which the linear
transformation is defined. We often oversee this requirement when using
the canonic basis, but we must be careful otherwise.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display, clear_output, Markdown
```

Here is an interactive example where you can change the coordinates of two vectors to form a new basis. The canonical basis is shown in red.

```python
base_original = np.array([[1, 0], [0, 1]])

vetor = np.array([3, 2])

def plot_base_transform(a1, a2, b1, b2):
    base_nova = np.array([[a1, a2], [b1, b2]])
    
    plt.figure(figsize=(5,5))
    plt.quiver(0, 0, base_original[0, 0], base_original[0, 1], angles='xy', scale_units='xy', scale=1, color='r', label='Canonical basis')
    plt.quiver(0, 0, base_original[1, 0], base_original[1, 1], angles='xy', scale_units='xy', scale=1, color='r')
    
    plt.quiver(0, 0, base_nova[0, 0], base_nova[0, 1], angles='xy', scale_units='xy', scale=1, color='b', label='New basis')
    plt.quiver(0, 0, base_nova[1, 0], base_nova[1, 1], angles='xy', scale_units='xy', scale=1, color='b')
    
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.grid()
    plt.legend()
    plt.show()

interact(plot_base_transform, 
         a1=(-3, 3, 1), a2=(-3, 3, 1), 
         b1=(-3, 3, 1), b2=(-3, 3, 1));
```

Note that for the vectors to constitute a valid basis, they must be linearly independent of each other (one vector cannot be a linear combination of the others; in this case, one cannot be a scalar multiple of the other). Consequently, in the plane, for the vectors to form a valid basis, they cannot be in the same direction. If you notice that the kernel stops responding, re-execute the cell.
