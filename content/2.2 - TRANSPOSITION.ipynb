## Transposition

For each linear transformation
$\mathbf{A}:\mathcal{U}\rightarrow\mathcal{W}$, we may construct another
linear transformation $\mathbf{A}^{\rm T}$ such that its representation
is the transposition of matrix $\mathbf{A}$. The element in the position
$(i,j)$ of a matrix $\mathbf{A}$ will become the element in position
$(j,i)$ of $\mathbf{A}^{\rm T}$.
$$\mathbf{A}^{\rm T}=\left[\begin{matrix}
a_{11} & a_{21} & \cdots & a_{m1}\\
a_{12} & a_{22} & \cdots & a_{m2}\\
\vdots & \vdots & \ddots & \vdots\\
a_{1n} & a_{2n} & \cdots & a_{mn}
\end{matrix}\right]$$

The transpose of a composite is the composite of the transposes in
reverse order:
$(\mathbf{B}\circ\mathbf{A})^{\rm T}=\mathbf{A}^{\rm T}\circ\mathbf{B}^{\rm T}$.

When a matrix $\mathbf{A}$ is equal to its transpose, i.e.,
$\mathbf{A}=\mathbf{A}^{\rm T}$, we call it a *symmetric* matrix. In the
particular case of symmetric matrices, i.e., when
$\mathbf{A}=\mathbf{A}^{\rm T}$, the set $\mathcal{R}(\mathbf{A})$ is
also equal to the set formed by all possible linear combinations of the
rows of matrix $\mathbf{A}$.

It will often be the case that we will refer to matrices and linear
transformations interchangeably. For the sake of simplicity, albeit
lacking some formality, we may refer to the range of a matrix when we
want to refer to the range of a function which is a linear
transformation induced, or represented, by a matrix.

Here is an example where we display the original image and the image representing its corresponding transposed matrix. Notice that during the transposition process, the first row becomes the first column, the second row becomes the second column, and so on. It might be intuitive to think that this process rotates the image, however, that is not what happens. What do you think this happens in this process?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_path = 'venetianmasks.png'  
image = mpimg.imread(image_path)

plt.figure(figsize=(6, 6))
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')
plt.show()

transposed_image = np.transpose(image, (1, 0, 2))

plt.figure(figsize=(6, 6))
plt.title("Transposed Image")
plt.imshow(transposed_image)
plt.axis('off')
plt.show()
```

Here is an example of a 3x3 matrix and its transpose. This program calculates the transpose of any 3x3 matrix. You can change the values of the variables that make up the matrix within the cell, and upon running it, observe the result.

```python
def transpose_matrix():

    a11, a12, a13 = 1, 2, 3
    a21, a22, a23 = 4, 0, 1
    a31, a32, a33 = 5, 2, 3
    
    matrix = np.array([[a11, a12, a13], 
                       [a21, a22, a23], 
                       [a31, a32, a33]])
    
    transposed_matrix = matrix.T
    
    print("Original Matrix:")
    print(matrix)
    print("\nTransposed Matrix:")
    print(transposed_matrix)

transpose_matrix()
```
