# Vector Spaces

From our field of choice $\mathbb{F}$, we may draw elements to construct
*vectors*, which may also be collected in a set. We are particularly
interested in sets whose elements (vectors) can be added and multiplied
by scalars $x\in\mathbb{F}$, and also satisfy some conditions related to
these two operations. Let us call this set a *vector space* over the
field $\mathbb{F}$ and the conditions to be satisfied the *vector space
properties*. Therefore, given a vector space $\mathcal{E}$ over the
field $\mathbb{F}$ and any two elements of this vector space, say
$\mathbf{u},\ \mathbf{v}\in\mathcal{E}$, we shall guarantee that
$\mathbf{u}+\mathbf{v}\in\mathcal{E}$ and $x\mathbf{u}\in\mathcal{E}$,
$x\in\mathbb{F}$. The properties associated with vector addition are:

-   Associativity: For any elements
    $\mathbf{u},\ \mathbf{v},\ \mathbf{w}\in\mathcal{E}$,
    $(\mathbf{u}+\mathbf{v})+\mathbf{w}=\mathbf{u}+(\mathbf{v}+\mathbf{w})$.

-   Commutativity: For any elements
    $\mathbf{u},\ \mathbf{v}\in\mathcal{E}$,
    $\mathbf{u}+\mathbf{v}=\mathbf{v}+\mathbf{u}$.

-   Identity: $\exists 0\in\mathcal{E}$ such that
    $0+\mathbf{u}=\mathbf{u}$ for any element
    $\mathbf{u}\in\mathcal{E}$.

-   Inverse: Any element $\mathbf{u}\in\mathcal{E}$ has a symmetric
    $-\mathbf{u}$ such that $\mathbf{u}+(-\mathbf{u})=0$.

The properties associated with vector multiplication by a scalar are:

-   Associativity: For any element $\mathbf{u}\in\mathcal{E}$ and
    scalars $x,\ y\in\mathbb{F}$, $(xy)\mathbf{u}=x(y\mathbf{u})$.

-   Identity: $\exists 1\in\mathbb{F}$ such that
    $1\mathbf{u}=\mathbf{u}$ for any element $\mathbf{u}\in\mathcal{E}$.

The operations addition and multiplication by a scalar also relate by
means of properties:

-   Distributivity of the multiplication by a scalar with respect to the
    vector addition: For any $x\in\mathbb{F}$ and
    $\mathbf{u},\ \mathbf{v}\in\mathcal{E}$,
    $x(\mathbf{u}+\mathbf{v})=x\mathbf{u}+x\mathbf{v}$.

-   Distributivity of the scalar addition with respect to the
    multiplication by a scalar: For any $x,\ y\in\mathbb{F}$ and
    $\mathbf{u}\in\mathcal{E}$,
    $(x+y)\mathbf{u}=x\mathbf{u}+y\mathbf{u}$.

We may build spaces consisting of different sorts of elements, such as
numbers, functions, or sets. However, we are particularly interested in
this book in spaces whose elements are vectors, although we do not need
to restrict our concept of vectors to any particular form at this
moment. We may consider, as an example, the vector space $\mathcal{E}$
whose elements are $n$-tuples $\mathbf{v}$ of scalars
$v_i\in\mathbb{R}$, displayed as follows: $$\mathbf{v}=\left[\begin{matrix}v_1\\v_2\\\vdots\\v_n\end{matrix}\right]$$
or, equivalently, $$
\mathbf{v}=\left[v_1\ v_2\ \cdots\ v_n\right]^{\rm T}$$ where the
superscript $[\cdot]^{\rm T}$ denotes the operation of *transposition*.

The reader may want to pause and verify that this is indeed a vector
space, for it satisfies all conditions stated above. This particular
vector space, $\mathcal{E}$, will be denoted here as $\mathbb{R}^n$.

Based on the properties just outlined, it may be already clear that the
origin, $\mathbf{v}=\mathbf{0}$, must belong to all vector spaces.
Therefore any other vector $\mathbf{v}$ as
in (<a href="#eq:vector" data-reference-type="ref"
data-reference="eq:vector">[eq:vector]</a>) may be regarded as a line
segmented extending from the origin to the point represented by the
$n$-tuple $(v_1,\ v_2,\ \ldots,\ v_n)$. The properties also lead
naturally to the definition of linear combination, outlined below:

A vector $\mathbf{v}\in\mathcal{E}$ defined over the field $\mathbb{F}$
is a *linear combination* of a collection of vectors
$\{\mathbf{v}_i\}\subseteq\mathcal{E}$, $i=1,\ \ldots,\ r$, if we can
write $\mathbf{v}$ as
$$\mathbf{v}=\sum_{i=1}^{r}\alpha_i\mathbf{v}_i,\quad\alpha_i\in\mathbb{F}$$

The definition of *linear independence* follows.

Given a subset $\mathcal{V}=\{\mathbf{v}_i\}\subseteq\mathcal{E}$,
$i=1,\ 2,\ \ldots,\ r$, defined over a field $\mathbb{F}$, we say that
the vectors $\mathbf{v}_i$ are *linearly independent* if
$$\alpha_1\mathbf{v}_1+\alpha_2\mathbf{v}_2+\cdots+\alpha_r\mathbf{v}_r=0,\quad\alpha_i\in\mathbb{F}$$
implies $\alpha_1=\alpha_2=\cdots=\alpha_r=0$. Conversely, we also say
that if the only linear combination of $\mathbf{v}_i$,
$i=1,\ \ldots,\ r$, equal to zero is the one for all coefficients equal
to zero, then the vectors $\mathbf{v}_i$ are linearly independent.

## Subspaces

Once we have defined vector spaces, we may now define vector subspaces
over a field as any subset of a given space for which the vector-space
conditions for addition and multiplication by a scalar apply.

\[@Hoff1971\]Let $\mathcal{E}$ be a vector space over the field
$\mathbb{F}$. The intersection of subspaces of $\mathcal{E}$ is also a
subspace of $\mathcal{E}$.

*Proof.* Let $\{\mathcal{V}_i\}$ be a collection of subspaces of
$\mathcal{E}$ and let $\mathcal{V}=\underset{i}{\cap}\mathcal{V}_i$.
Then the elements of $\mathcal{V}$ belong to each and every subspace
$\mathcal{V}_i$. As any two vectors
$\mathbf{u},\ \mathbf{v}\in\mathcal{V}$ also belong to each and every
$\mathcal{V}_i$ and as each and every $\mathcal{V}_i$ is a vector space
itself, then $\mathbf{u}+\mathbf{v}$ belongs to each and every
$\mathcal{V}_i$ and, consequently, also belongs to $\mathcal{V}$. A
similar reasoning can be applied to the multiplication by a scalar.

As it was seen before, a vector space must allow sum of vectors and multiplication by a scalar. Also, by definition, the vector spaces must be close, that is, the result of said operations must belong to the same vector space as the operands.

```python
from PIL import Image 
import numpy as np
from matplotlib import pyplot as plt
import ipywidgets as widgets
```

The example below is a typical representation of the $R_2$ vector space. A regular cartesian coordinates plane, in which any vector $v$ is represented in the form of:
$$\mathbf{v}=\left[\begin{matrix}x\\y\end{matrix}\right]$$
In which $x$ represents the horizontal coordinate adn $y$ represents the vertical coordinate.
Notices that the blue vector is the sum of the black and red vectors. This symbolizes vector sum and how it is a closed opperation.

```python
escalar01 = widgets.FloatText(value = 1, description = 'x of red:', disabled = False) 
escalar02 = widgets.FloatText(value = 2, description = 'y of red:', disabled = False)
escalar03 = widgets.FloatText(value = 2, description = 'x of black:', disabled = False)
escalar04 = widgets.FloatText(value = 1, description = 'y of black:', disabled = False)
formatted1 = widgets.VBox([escalar01, escalar02])
formatted2 = widgets.VBox([escalar03, escalar04])
formatted3 = widgets.HBox([formatted1, formatted2])

def vecSpace(escalar01, escalar02, escalar03, escalar04):
    plt.figure(figsize=(2, 2))
    plt.arrow(0, 0, escalar01, escalar02, width = 0.01, color = 'red', length_includes_head = True, head_width = 0.03)
    plt.arrow(0, 0, escalar03, escalar04, width = 0.01, color = 'black', length_includes_head = True, head_width = 0.03)
    plt.arrow(0, 0, escalar01+escalar03, escalar02+escalar04, width = 0.01, color = 'blue', length_includes_head = True, head_width = 0.03)
    plt.grid()
    plt.show()
    return 0
    
verif = widgets.interactive_output(vecSpace, {'escalar01': escalar01, 'escalar02': escalar02, 
                                           'escalar03': escalar03, 'escalar04': escalar04})
display(formatted3, verif)
```

It is also possible to observe properties of vector spaces in the context of image processing. As it was said in the previous section, an image can be represented in matrix form, in which the numerical value of each element represents the color of the pixel. Also in the previous section, we also learned that, in order to represent a grayscale image with the best computational efficiency, the values that a pixel can assume are integers that range from 0 to 255. Its is defined that the highest the value, the closest it is to white. Below is a simple grayscale image.

```python
earthImage = Image.open("earth_gs.jpg")
plt.axis('off')
plt.imshow(earthImage, cmap='gray')
```

If we wish to make this image darker and duller, it would be logical to make is values closer to 0, since 0 represents pure black. So, by multiplying the image's array by 0.5 and recontructing it, we would be able to see an image without such bright colors. Once again, this symbolizes the vector space operation of scalar multiplication. Notice a very important mathematical detail, however. A scalar that multiplies any element of a vector space must belong to the same field with which the vector space is defined. 0.5 does not belong to the integers that range from 0 to 255. So, it is necessary to use the real numbers (ranging from 0 to 1), in order to represent the image with mathematical coherence. That will be done in the following cell. Notice that, before plotting the image, we multiply once more by 256. It is necessary to do so for the plotting function to work correctly.

```python
earthArray = 1/256*np.array(earthImage)
earthArray = 0.5*earthArray
earthMult = Image.fromarray(256*earthArray)
plt.axis('off')
plt.imshow(earthMult, cmap='gray')
```
